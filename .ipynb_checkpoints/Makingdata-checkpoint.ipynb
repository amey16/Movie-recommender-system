```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x): # is the movie_title
    genres = []
    try:
        result = tmdb_movie.search(x) #getting that movie
    except TypeError:
        return np.NaN
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json() #making json of genres
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres) #return string of genres joined by spaces
    else:
        np.NaN
```

```python
df_american['genre'] = df_american['Title'].apply(get_genre)
df_american
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_american.columns
```

```python
df_american = df_american[['Title','Cast and crew','genre']]
```

```python
df_american['Director'] = df_american['Cast and crew'].apply(get_director)
```

```python
def get_actor1(x):
    return (x.split("screenplay); ")[-1].split(", ")[0])
```

```python
#in 2018movies bollywood we cannot have actors as they are joined together we will be only
#using the american movies dataset for 2018 movies
```

```python
def get_actor2(x):
    if len(x.split("screenplay); ")[-1].split(", ")) <2 :
        return np.NaN
    else:
        return (x.split("screenplay); ")[-1]).split(", ")[1]
```

```python
def get_actor3(x):
    if len(x.split("screenplay); ")[-1].split(", ")) <3:
        return np.NaN
    else:
        return (x.split("screenplay); ")[-1]).split(", ")[2]
```

```python
df_american['actor_1_name'] = df_american['Cast and crew'].map(lambda x: get_actor1(x))
df_american['actor_2_name'] = df_american['Cast and crew'].map(lambda x: get_actor2(x))
df_american['actor_3_name'] = df_american['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
df_american.columns
```

```python
df_american = df_american.rename(columns={'Title':'movie_title'})
df_american = df_american.rename(columns={'Director':'director_name'})
df_american = df_american.rename(columns={'genre':'genres'})
```

```python
df_american = df_american.drop(columns = ['Cast and crew'])
df_american
```

```python
new_data = df_american.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
new_data['actor_2_name'] = new_data['actor_2_name'].replace(np.nan, 'unknown')
new_data['actor_3_name'] = new_data['actor_3_name'].replace(np.nan, 'unknown')
new_data['movie_title'] = new_data['movie_title'].str.lower()
new_data['comb'] = new_data['actor_1_name'] + ' ' + new_data['actor_2_name'] + ' '+ new_data['actor_3_name'] + ' '+ new_data['director_name'] +' ' + new_data['genres']
```

# For 2020

```python
import bs4 as bs
import urllib.request
```

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2020"
source = urllib.request.urlopen(link).read()
soup = bs.BeautifulSoup(source,'lxml')
```

```python
tables = soup.find_all('table',class_='wikitable sortable') #movies separted into quaters
```

```python
df1 = pd.read_html(str(tables[0]))[0]
df2 = pd.read_html(str(tables[1]))[0]
df3 = pd.read_html(str(tables[2]))[0]
df4 = pd.read_html(str(tables[3]).replace("'1\"\'",'"1"'))[0] # avoided "ValueError: invalid literal for int() with base 10: '1"'
```

```python
df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
df
```

```python
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    if not result:
        return np.NaN
    else:
        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        data_json = response.json()
        if data_json['genres']:
            genre_str = " " 
            for i in range(0,len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
        else:
            return np.NaN
```

```python
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
```

```python
df_american = df[['Title','Cast and crew','genres']]
df_american
```

```python
df_american['director_name'] = df_american['Cast and crew'].map(lambda x: get_director(str(x)))
df_american['actor_1_name'] = df_american['Cast and crew'].map(lambda x: get_actor1(x))
df_american['actor_2_name'] = df_american['Cast and crew'].map(lambda x: get_actor2(x))
df_american['actor_3_name'] = df_american['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
df_american = df_american.rename(columns={'Title':'movie_title'})
new_data2 = df_american.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
new_data2['actor_2_name'] = new_data2['actor_2_name'].replace(np.nan, 'unknown')
new_data2['actor_3_name'] = new_data2['actor_3_name'].replace(np.nan, 'unknown')
new_data2['movie_title'] = new_data2['movie_title'].str.lower()
new_data2['comb'] = new_data2['actor_1_name'] + ' ' + new_data2['actor_2_name'] + ' '+ new_data2['actor_3_name'] + ' '+ new_data2['director_name'] +' ' + new_data2['genres']
```
