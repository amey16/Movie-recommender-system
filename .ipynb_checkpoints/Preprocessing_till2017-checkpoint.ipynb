```python
#replacing all the null values
data['actor_1_name'] = data['actor_1_name'].replace(np.nan,'unknown')
data['actor_2_name'] = data['actor_2_name'].replace(np.nan,'unknown')
data['actor_3_name'] = data['actor_3_name'].replace(np.nan,'unknown')
data['director_name'] = data['director_name'].replace(np.nan,'unknown')
```

```python
def replace(x):
    x = x.replace("|"," ")
    return x
data['genres'] = data['genres'].apply(replace)
data['movie_title'] = data['movie_title'].str.lower()
```

```python
#\xa0 represents the end of string which we can remove so -
data['movie_title'] = data['movie_title'].apply(lambda x:x[:-1])
```

```python
credits = pd.read_csv('datasets/credits.csv')
credits
```

```python
meta = pd.read_csv('datasets/movies_metadata.csv')
```

```python
meta.head()
```

```python
meta['release_date'] = pd.to_datetime(meta['release_date'] ,errors = 'coerce')
meta['year'] = meta['release_date'].dt.year
```

```python
meta['year'].value_counts().sort_index()
```

```python
#we want data after 2016 so taking only year 2017 data
new_meta = meta.loc[meta.year == 2017][['genres','id','title','year']]
new_meta
```

```python
new_meta['id'] = new_meta['id'].astype(int)
```

```python
data1 = pd.merge(new_meta,credits, on="id")
data1
```

```python
from ast import literal_eval as le
#converting string into list using literaleval
data1['genres'] = data1['genres'].apply(lambda x:le(x))
data1['cast'] = data1['cast'].apply(lambda x:le(x)) 
data1['crew'] = data1['crew'].apply(lambda x:le(x))
```

```python
def genre_engineering(x):
    string=" "
    gen = []
    for i in x:
        if i.get('name') == 'Science Fiction':
            gen.append('Sci-Fi')
        else:
            gen.append(i.get('name'))
    if(gen == []):
        return np.NaN
    else:
        return string.join(gen)
```

```python
data1['genres'] = data1['genres'].apply(genre_engineering)
data1['genres']
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if(casts == []):
        return np.NaN
    else:
        return casts[0]
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if(casts == [] or len(casts)<=1):
        return np.NaN
    else:
        return casts[1]
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if(casts == [] or len(casts)<=2):
        return np.NaN
    else:
        return casts[2]
```

```python
def get_directors(x):
    direct = []
    s = " "
    for i in x:
        if i.get('job') == 'Director':
            direct.append(i.get('name'))
    if(direct == []):
        return np.NaN
    else:
        return s.join(direct)
```

```python
data1['actor_1_name'] = data1['cast'].map(lambda x:get_actor1(x))
data1['actor_2_name'] = data1['cast'].map(lambda x:get_actor2(x))
data1['actor_3_name'] = data1['cast'].map(lambda x:get_actor3(x))
data1['director_name'] = data1['crew'].map(lambda x:get_directors(x))
```

```python
data1.columns
```

```python
data1 = data1.loc[:][['director_name','actor_1_name','actor_2_name','actor_3_name','genres','title']]
data1
```

```python
data1 = data1.dropna(how = "any")
```

```python
data1 = data1.rename(columns = {'title':'movie_title'})
```

```python
data1.head(10)
```

```python
data1['movie_title'] = data1['movie_title'].str.lower()
data1['combined'] = data1['actor_1_name'] + ' ' + data1['actor_2_name'] + ' ' + data1['actor_2_name']+ ' ' + data1['director_name'] + ' ' +data1['genres']
```
